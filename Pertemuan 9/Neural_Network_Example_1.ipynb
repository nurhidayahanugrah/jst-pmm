```python
# Tabel logika AND 2 input

"""
x1  x2  y
1   1   1
1   0   0
0   1   0
0   0   0
"""
```

```python
import keras
import numpy as np
from keras import backend as K

# model
model = keras.Sequential()
model.add(keras.layers.Dense(units=1, input_shape=[2]))
model.add(keras.layers.Dense(units=5, name='hidden1', activation="softmax"))
model.add(keras.layers.Dense(units=7, name='hidden2', activation="relu"))
model.add(keras.layers.Dense(units=5, name='hidden3', activation="softmax"))
model.add(keras.layers.Dense(units=1, name='output'))

# model compile
model.compile(optimizer='adam', loss='mean_squared_error')

K.set_value(model.optimizer.learning_rate, 0.001)

# set data
xs = np.array([[1,1], [1,0], [0,1], [0,0]], dtype=int)
ys = np.array([1, 0, 0, 0], dtype=int)
```

```python
# tampilkan arsitektur awal

model.summary()
```

```python
# cek bobot model awal
weights = model.get_weights()
weights
```

```python
# training
model.fit(xs, ys, epochs=1000)
```

```python
# test
data = [1,1]
answer = model.predict([data])
print(answer)
```
