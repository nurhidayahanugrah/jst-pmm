```python
import numpy as np
import keras
from keras import backend as K

# model
model2 = keras.Sequential()
model2.add(keras.layers.Dense(units=1, input_shape=[2]))
model2.add(keras.layers.Dense(units=7, activation="elu", name="Hidden1"))
model2.add(keras.layers.Dense(units=7, activation="relu", name="Hidden2"))
model2.add(keras.layers.Dense(units=3, activation="softmax", name="Hidden3"))
model2.add(keras.layers.Dense(units=1, name="output"))

# model compile
model2.compile(optimizer='sgd', loss='mean_squared_error')

K.set_value(model2.optimizer.learning_rate, 0.001)

# set data
xs = np.array([[2, 3], [4, 1], [5, 4], [7, 5], [8, 2], [2,1], [4,9], [8,2], [7,1], [6,5], [1,1], [3, 2]], dtype=int)
ys = np.array([16, 12, 28, 34, 24, 8, 44, 24, 18, 32, 6, 14], dtype=int)

model2.summary()
```

```python
# cek bobot model awal
weights = model2.get_weights()
weights
```

```python
# training

model2.fit(xs, ys, epochs=500)
```

```python
# test
data = [3,2]
answer = model2.predict([data])
print(answer)
```
