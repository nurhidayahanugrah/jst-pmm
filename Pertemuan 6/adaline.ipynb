```python
def ada_train(x, t, alpha=.1, max_err=.1, draw=False):
  w = np.random.uniform(0, 1, len(x[0]) + 1)
  b = np.ones((len(x), 1))
  x = np.hstack((b, x))
  stop = False
  epoch = 0

  while not stop:
    epoch += 1
    max_ch = -sys.maxsize

    print('\nEpoch #%d' % epoch)

    for r, row in enumerate(x):
      y = np.dot(row, w)

      for i in range (len(row)):
        w_new = w[i] + alpha * (t[r] - y) * row[i]
        max_ch = max(abs(w[i] - w_new), max_ch)
        w[i] = w_new

      print('Bobot: {}'.format(w))

      if draw:
        plot(line(w), x, t)

        stop = max_ch < max_err

  return w

# Logika AND
# Input bipolar, output bipolar

train = [[1, 1], [1, -1], [-1, 1], [-1, -1]]
target = [1, -1, -1, -1]
w = ada_train(train, target, draw=True)
```
